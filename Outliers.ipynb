```python
import numpy as np

arr = np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
arr
```

```python
Q1 = np.percentile(arr, 25)
Q2 = np.percentile(arr,50)
Q3 = np.percentile(arr,75)
```

```python
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
upper = Q3 + 1.5*(IQR)
upper
```

```python
lower = Q1 - 1.5*(IQR)
lower
```

```python
l = []
for i in arr:
  if i >= lower and i <= upper:
    l.append(i)

l
```
